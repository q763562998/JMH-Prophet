```python
import pickle

def save_dict_to_disk(data, filename):
    with open(filename, 'wb') as file:
        pickle.dump(data, file)

def load_dict_from_disk(filename):
    with open(filename, 'rb') as file:
        loaded_data = pickle.load(file)
    return loaded_data
```

```python
nb = 0
labels = []
data_set = []
for k,val in new_tree.items():
    edge_attributes = [my_onehot(x[0]) for x in val[0][2]]
    edge_list_tmp = val[0][1]

    edge_list = []
    c = 0
    for u,v in np.array(edge_list_tmp).transpose():
        edge_list.append((u,v,{"x":edge_attributes[c]}))
        c = c + 1  
    g = nx.from_edgelist(edge_list)
    # compute node feature
    for node in g.nodes():
        tmp = np.zeros(13)
        for edge in g.edges(node):
            tmp = tmp + g.edges()[edge]["x"]

        g.nodes()[node]["x"] = tmp
    # print(y_dict)
    
    # k=k+".java"
    data_set.append(k)
    print(k)
    
    # nx.write_gpickle(g,"Datasets/graphs/zhs/setandteardown/"+str(nb)+".pk")
    nx.write_gpickle(g,"/home/zhs/colefunda/zhs/model/data/pk/eclipse080001/"+k+".pk")
    # labels.append([nb,y_dict[k]])
    nb = nb + 1
    if nb % 10 == 0:
        print(nb)
```
